```python
# Create a vector of a given norm in a direction of a given vector
def enorm(x, module):
    return x * (module / norm(x))
```

### Exercise 2.6

Illustrate that norm of the vector is equal to square root of the dot of the vector with itself

Well, the logic is quite clear. The norm of the vector is:
$$norm = \sqrt{\sum_{i=1}^{n} x_i^2}$$

Whereas square root of the dot of x with itself is:

$$dot = \sqrt{\sum_{i=1}^{n} x_i * x_i}$$

```python
a = vec()

norm(a), np.dot(a,a)**0.5
```

### Exercise 2.9

Define orthogonal projection and plot examples

```python
def projection(projected, base):
    return base*np.dot(projected, base)/np.dot(base,base)
```

```python
base = vec()
projected = vec()
```

```python
plt.arrow(0,0, *base, head_width=.3, width=.1, length_includes_head=True, color='brown')
plt.arrow(0,0, *projected, head_width=.3, width=.1, length_includes_head=True, color = 'blue')
plt.arrow(0,0, *projection(projected, base), head_width=.3, length_includes_head=True, color = 'blue', linestyle='--')
second_projection = projected - projection(projected, base)
plt.arrow(0,0, *second_projection, head_width=.3,  length_includes_head=True, color = 'blue', linestyle='--')

plt.arrow(*projection(projected, base), *second_projection, head_width=0, color = 'blue',linestyle=':')
plt.arrow(*second_projection, *(projection(projected, base)),  head_width=0, color = 'blue',linestyle=':')
plt.plot(3,3, 5,5)
plt.axis('square')
plt.axis([-5,10,-5, 10])
```
