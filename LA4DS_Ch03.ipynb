### Exercise 3.1

Write a function for linear combination of vectors

```python
def lincomb(scalars, vectors):
    return sum([i*j for i, j in zip(scalars, vectors)])
```

```python
a = vec()
b = vec()
c = vec()

scalars = np.random.randint(-5,5, size=3)

x = lincomb(scalars, [a,b,c])
```

```python
scalars
```

```python
np.array([a,b,c]).T @ np.array(scalars).T, lincomb(scalars, [a,b,c])
```

```python
a_l = a*scalars[0]
b_l = b*scalars[1]
c_l = c*scalars[2]



plt.arrow(0,0, *a_l, color = 'green', length_includes_head=True, width=0.1, head_width=0.3)
plt.arrow(*a_l, *b_l, color = 'green', length_includes_head=True, width=0.1, head_width=0.3)
plt.arrow(*(a_l+b_l), *c_l, color = 'green', length_includes_head=True, width=0.1, head_width=0.3)
plt.arrow(0,0, *x, color = 'orange', length_includes_head=True, width=0.1, head_width=0.3)

plt.axvline(0,0, color='grey', ls='--')
plt.axhline(0,0, color='grey', ls='--')
```

### Exercise 3.2

Illustrate how different spaces (R2, R3) behave

```python
# Case for R2

lambdas = np.array([np.random.uniform(-4, 4, 100)])

x = np.array([[1,3]])

dots = x.T @ lambdas
```

```python
sns.scatterplot(x=dots[0], y=dots[1])
```

```python
# R3 cases: 2 vectors 

lambdas = np.array(np.random.uniform(-4, 4, 200)).reshape((2,100))

a = np.array([3,5,1])
b = np.array([0,2,2])

points = np.array([a,b]).T @ lambdas
```

```python
go.Figure(data=[go.Scatter3d(
                        x=points[0, :],
                        y=points[1, :],
                        z=points[2, :],
                        mode='markers')])

# all points belong to one plane
```

```python
# R3 case 3 vectors

lambdas = np.array(np.random.uniform(-4, 4, 300)).reshape((3,100))

a = vec(3)
b = vec(3)
c = vec(3)

points = np.array([a, b, c]).T @ lambdas
```

```python
# Check independency

np.linalg.det(np.array([a,b,c])) # if independent, plot the points
```

```python
go.Figure(data=[go.Scatter3d(
                        x=points[0, :],
                        y=points[1, :],
                        z=points[2, :],
                        mode='markers')])

# Some random dots in R3 neither in one plane, nor on one line
```

```python
# R3 case with non-independent vectors

a = vec(3)
b = 0.5*a

lambdas = np.array(np.random.uniform(-4, 4, 200)).reshape((2,100))

points = np.array([a, b]).T @ lambdas
```

```python
go.Figure(data=[go.Scatter3d(
                        x=points[0, :],
                        y=points[1, :],
                        z=points[2, :],
                        mode='markers')])

# The thing is clearly a line
```
