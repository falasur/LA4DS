### Exercise 3.1

Write a function for linear combination of vectors

```python
def lincomb(scalars, vectors):
    return sum([i*j for i, j in zip(scalars, vectors)])
```

```python
a = vec()
b = vec()
c = vec()

scalars = np.random.randint(-5,5, size=3)

x = lincomb(scalars, [a,b,c])
```

```python
scalars
```

```python
np.array([a,b,c]).T @ np.array(scalars).T, lincomb(scalars, [a,b,c])
```

```python
a_l = a*scalars[0]
b_l = b*scalars[1]
c_l = c*scalars[2]



plt.arrow(0,0, *a_l, color = 'green', length_includes_head=True, width=0.1, head_width=0.3)
plt.arrow(*a_l, *b_l, color = 'green', length_includes_head=True, width=0.1, head_width=0.3)
plt.arrow(*(a_l+b_l), *c_l, color = 'green', length_includes_head=True, width=0.1, head_width=0.3)
plt.arrow(0,0, *x, color = 'orange', length_includes_head=True, width=0.1, head_width=0.3, ls=':')

plt.axvline(0,0, color='grey', ls='--')
plt.axhline(0,0, color='grey', ls='--')
```
